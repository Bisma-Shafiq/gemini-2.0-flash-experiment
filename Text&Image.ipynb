```python
# API Key setup
from google.colab import userdata
GOOGLE_API_KEY: str = userdata.get('GOOGLE_API_KEY')
if(GOOGLE_API_KEY):
  print("API Key found")
else:
  print("API Key not found")
```

```python
from google import genai
from google.genai import Client
client: Client = genai.Client(api_key=GOOGLE_API_KEY)
```

```python
from google.genai.types import GenerateContentResponse
from IPython.display import display, Markdown
```

# Text

```python
model : str = "gemini-2.0-flash-exp"
response:GenerateContentResponse= client.models.generate_content(model = model ,contents ="what is ai in 10 words?")
```

```python
display(response.text)
```

# Image

```python
# Display the downloaded image
import PIL.Image
img = PIL.Image.open('/content/Black Textured Background Twitter Post (1).png')
display(img)
```

```python
prompt = {"Explain the image. and write a gig description for me so i add it on freelance platform."}
```

```python
response : GenerateContentResponse = client.models.generate_content(model=model, contents = [img,prompt])
display(Markdown(response.text))
```
